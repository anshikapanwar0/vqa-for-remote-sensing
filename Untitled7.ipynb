```python
import os
import cv2
import matplotlib.pyplot as plt

# Directory containing the images
RSVQAimages_dir = '/content/drive/MyDrive/Images_LR'

image_paths = []

# Traverse the directories and get all .tif file paths
for root, dirs, files in os.walk(RSVQAimages_dir):
    for file in files:
        if file.endswith('.tif'):
            full_path = os.path.join(root, file)
            image_paths.append(full_path)

# Check dimensions and consistency across images
def check_image_dimensions(image_paths):
    dimensions = []

    for idx, path in enumerate(image_paths):
        print(f"Loading image: {path}")

        # Load the image
        img = cv2.imread(path, cv2.IMREAD_UNCHANGED)

        if img is None:
            print(f"Error loading image: {path}")
            continue

        # Get image dimensions
        height, width = img.shape[:2]
        print(f"Image Dimensions (HxW): {height} x {width}")
        dimensions.append((height, width))

        # Visualize the first 5 images
        if idx < 5:
            plt.figure(figsize=(5, 5))
            plt.imshow(img, cmap='gray')
            plt.title(f"Image {idx+1}: {os.path.basename(path)}")
            plt.axis('off')
            plt.show()

    return dimensions

# Check dimensions of images
image_dimensions = check_image_dimensions(image_paths)

unique_dimensions = set(image_dimensions)
print(f"\nUnique image dimensions found: {unique_dimensions}")

# Save image paths to a text file
with open('image_paths.txt', 'w') as f:
    for path in image_paths:
        f.write(path + '\n')
```

```python
import os
import cv2

# Directory containing the images
RSVQAimages_dir = '/content/drive/MyDrive/Images_LR'
resized_images_dir = '/content/drive/MyDrive/Images_LR_resized'

# Create the resized images directory if it doesn't exist
if not os.path.exists(resized_images_dir):
    os.makedirs(resized_images_dir)

def resize_images(image_paths, target_size=(384, 384)):
    for path in image_paths:
        print(f"Processing image: {path}")

        # Load the image
        img = cv2.imread(path, cv2.IMREAD_UNCHANGED)

        if img is None:
            print(f"Error loading image: {path}")
            continue

        # Resize the image
        resized_img = cv2.resize(img, target_size, interpolation=cv2.INTER_LINEAR)

        # Construct new file path
        base_name = os.path.basename(path)
        new_path = os.path.join(resized_images_dir, base_name)

        # Save the resized image
        cv2.imwrite(new_path, resized_img)
        print(f"Saved resized image to: {new_path}")

# Resize and save the images
resize_images(image_paths)
```

```python
import json
import pandas as pd
import os
import torch
from torchvision import models, transforms
from PIL import Image

# Initialize the ResNet model and set it to evaluation mode
resnet_model = models.resnet50(pretrained=True)
resnet_model.eval()

# Define the image transformations
image_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Directory where images are stored
image_dir = '/content/drive/MyDrive/Images_LR_resized'
output_embeddings_file = 'image_embeddings_resnet.csv'

# List to store image embeddings
image_embeddings_list = []

# Process each image to compute embeddings
for img_filename in os.listdir(image_dir):
    if img_filename.endswith('.tif'):
        img_id = img_filename.split('.')[0]
        image_path = os.path.join(image_dir, img_filename)

        try:
            # Load and preprocess the image
            image = Image.open(image_path).convert('RGB')
            image_tensor = image_transforms(image).unsqueeze(0)

            # Pass the image through the ResNet model to get embeddings
            with torch.no_grad():
                embeddings = resnet_model(image_tensor).squeeze(0).tolist()

            # Store the embeddings with image ID
            image_embeddings_list.append({
                'img_id': img_id,
                'image_embeddings': embeddings
            })

        except Exception as e:
            print(f"Error processing image {img_filename}: {e}")

# Convert to DataFrame and save to CSV
image_embeddings_df = pd.DataFrame(image_embeddings_list)
image_embeddings_df.to_csv(output_embeddings_file, index=False)

print(f"Saved image embeddings for {len(image_embeddings_list)} images to {output_embeddings_file}.")
```

```python
import json
import pandas as pd
import os
from transformers import ViltProcessor
from PIL import Image

# Initialize the ViLT processor
vilt_processor = ViltProcessor.from_pretrained("dandelin/vilt-b32-mlm")

# Load questions from JSON file
with open('/content/drive/MyDrive/all_questions.json', 'r') as f:
    questions_data = json.load(f)

# Load answers from JSON file
with open('/content/drive/MyDrive/all_answers.json', 'r') as f:
    answers_data = json.load(f)

# Dictionary to map answer IDs to answers
answer_dict = {answer['id']: answer['answer'] for answer in answers_data['answers']}

# Directory where images are stored
image_dir = '/content/drive/MyDrive/Images_LR_resized'

# Batch size for processing
batch_size = 100
processed_data = []

def process_batch(start, end):
    batch_data = []
    for i, question in enumerate(questions_data['questions'][start:end], start=start):
        question_id = question['id']
        question_text = question['question']
        answers_ids = question['answers_ids']
        img_id = question['img_id']

        # Fetch corresponding answers using answer IDs
        answers_text = [answer_dict[ans_id] for ans_id in answers_ids if ans_id in answer_dict]

        # Construct the image path using img_id
        image_filename = f"{img_id}.tif"
        image_path = os.path.join(image_dir, image_filename)

        # Check if the image exists
        if os.path.exists(image_path):
            try:
                # Load and process the image
                image = Image.open(image_path).convert('RGB')
                encoding = vilt_processor(image, question_text, return_tensors="pt", padding=True)

                # Store processed data with input_ids and attention_mask
                batch_data.append({
                    'question_id': question_id,
                    'question_text': question_text,
                    'answers': answers_text,
                    'img_id': img_id,
                    'active': question['active'],
                    'input_ids': encoding['input_ids'].squeeze(0).tolist(),
                    'attention_mask': encoding['attention_mask'].squeeze(0).tolist()
                })
            except Exception as e:
                print(f"Error processing image {image_filename}: {e}")
        else:
            print(f"Image {image_filename} not found.")

    # Append the batch data to the global processed data list
    return batch_data

# Process all questions in batches
total_questions = len(questions_data['questions'])
for start in range(0, total_questions, batch_size):
    end = min(start + batch_size, total_questions)
    print(f"Processing batch {start} to {end}...")

    # Process the current batch
    batch_processed_data = process_batch(start, end)

    # Append to global processed data
    processed_data.extend(batch_processed_data)

    # Convert batch to DataFrame and save to CSV to avoid memory overload
    batch_df = pd.DataFrame(batch_processed_data)
    if start == 0:
        batch_df.to_csv('processed_qa_data.csv', index=False, mode='w')
    else:
        batch_df.to_csv('processed_qa_data.csv', index=False, mode='a', header=False)

print(f"Processed {len(processed_data)} questions with their corresponding answers.")

final_df = pd.read_csv('processed_qa_data.csv')
print(f"Final CSV saved with {len(final_df)} rows.")
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import ast

# Load the processed data from the CSV file
processed_df = pd.read_csv('processed_qa_data.csv')

# Initialize the Label Encoder
label_encoder = LabelEncoder()

# Flatten the list of answers to fit the label encoder
all_answers = [answer for sublist in processed_df['answers'] if pd.notnull(sublist) for answer in ast.literal_eval(sublist)]

# Fit the LabelEncoder and transform the answers
encoded_answers = label_encoder.fit_transform(all_answers)

# Create a dictionary to map original answers to their encoded values
answer_mapping = dict(zip(label_encoder.classes_, range(len(label_encoder.classes_))))

# Function to encode answers safely
def encode_answers(answer_list):
    if pd.isnull(answer_list):
        return []
    try:
        return [answer_mapping[ans] for ans in ast.literal_eval(answer_list) if ans in answer_mapping]
    except (ValueError, SyntaxError):
        return []

# Apply the label encoding to the 'answers' column
processed_df['encoded_answers'] = processed_df['answers'].apply(encode_answers)

processed_df.to_csv('processed_qa_data.csv', index=False)

print("Label encoding completed and saved to processed_qa_data.csv")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import ast
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import ViltProcessor, ViltForQuestionAnswering
from tqdm import tqdm

# Load the processed data from the CSV file
processed_df = pd.read_csv('processed_qa_data.csv')

# Initialize the Label Encoder
label_encoder = LabelEncoder()

# Flatten the list of answers to fit the label encoder
all_answers = [answer for sublist in processed_df['answers'] if pd.notnull(sublist) for answer in ast.literal_eval(sublist)]

# Fit the LabelEncoder and transform the answers
label_encoder.fit(all_answers)

# Create a dictionary to map original answers to their encoded values
answer_mapping = dict(zip(label_encoder.classes_, range(len(label_encoder.classes_))))

# Apply the label encoding to the 'answers' column
processed_df['encoded_answers'] = processed_df['answers'].apply(lambda x: [answer_mapping[ans] for ans in ast.literal_eval(x)])

processed_df.to_csv('processed_qa_data.csv', index=False)

image_embeddings_dict = {
    0: torch.rand(3, 384, 384),
    1: torch.rand(3, 384, 384)
}
processed_df['image_embedding'] = processed_df.index.map(image_embeddings_dict.get)

# Split data into train and test sets
train_df, test_df = train_test_split(processed_df, test_size=0.2, random_state=42)

# Custom Dataset for VQA
class VQADataset(Dataset):
    def __init__(self, dataframe, num_classes):
        self.dataframe = dataframe
        self.num_classes = num_classes

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        question = row['question_text']

        # Image embedding tensor
        image_embedding = row['image_embedding']
        image_embedding = torch.tensor(image_embedding).float() if image_embedding is not None else torch.zeros(3, 384, 384)

        # One-hot encoding for the target answers
        encoded_answers = row['encoded_answers']
        encoded_answer_tensor = torch.zeros(self.num_classes)
        for answer in encoded_answers:
            encoded_answer_tensor[answer] = 1

        return question, image_embedding, encoded_answer_tensor

# Initialize ViLT model and processor
processor = ViltProcessor.from_pretrained('dandelin/vilt-b32-mlm')
model = ViltForQuestionAnswering.from_pretrained('dandelin/vilt-b32-mlm')

# Set the number of output classes
num_classes = 1527  # Number of unique answers in your dataset
model.classifier = torch.nn.Linear(model.config.hidden_size, num_classes)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Initialize the datasets and dataloaders for training and testing
train_dataset = VQADataset(train_df, num_classes)
test_dataset = VQADataset(test_df, num_classes)

train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Define optimizer
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)

# Training loop
num_epochs = 3
model.train()
for epoch in range(num_epochs):
    for idx, (questions, image_embeddings, encoded_answers) in enumerate(tqdm(train_dataloader)):
        # Process questions
        processed_questions = processor(text=questions, images=image_embeddings, return_tensors="pt", padding=True, truncation=True).to(device)

        # Move tensors to device
        image_embeddings = image_embeddings.to(device)
        encoded_answers = encoded_answers.to(device)

        # Forward pass
        outputs = model(input_ids=processed_questions['input_ids'],
                        attention_mask=processed_questions['attention_mask'],
                        pixel_values=image_embeddings,
                        labels=encoded_answers)

        loss = outputs.loss

        # Backward pass and optimization steps
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if idx % 10 == 0:
            print(f"Epoch: {epoch}, Batch: {idx}, Loss: {loss.item()}")

# Save the trained model
model.save_pretrained("vilt_vqa_model")
processor.save_pretrained("vilt_vqa_model")
```

```python
import torch
from tqdm import tqdm
from transformers import ViltProcessor, ViltForQuestionAnswering
from torch.utils.data import DataLoader

# Load the saved model from the trained checkpoint
model = ViltForQuestionAnswering.from_pretrained("vilt_vqa_model")

# Load the saved processor from the checkpoint
processor = ViltProcessor.from_pretrained("vilt_vqa_model")

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Evaluation function to test the model on the test data
def evaluate_model(model, test_dataloader, processor, device):
    model.eval()
    total_loss = 0.0
    total_correct = 0
    total_samples = 0

    with torch.no_grad():
        for idx, (questions, image_embeddings, encoded_answers) in enumerate(tqdm(test_dataloader)):
            # Process questions and images
            processed_questions = processor(
                text=questions,
                images=image_embeddings,
                return_tensors="pt",
                padding=True,
                truncation=True
            ).to(device)

            # Move tensors to device
            image_embeddings = image_embeddings.to(device)
            encoded_answers = encoded_answers.to(device)

            # Forward pass
            outputs = model(
                input_ids=processed_questions['input_ids'],
                attention_mask=processed_questions['attention_mask'],
                pixel_values=image_embeddings,
                labels=encoded_answers
            )

            # Calculate loss
            loss = outputs.loss
            total_loss += loss.item()

            # Get predictions and calculate accuracy
            logits = outputs.logits
            predicted_answers = torch.argmax(logits, dim=-1)

            # One-hot encoding for comparison with encoded_answers
            target_answers = torch.argmax(encoded_answers, dim=-1)
            correct_predictions = (predicted_answers == target_answers).sum().item()
            total_correct += correct_predictions
            total_samples += encoded_answers.size(0)

    avg_loss = total_loss / len(test_dataloader)
    accuracy = total_correct / total_samples * 100

    print(f"Test Loss: {avg_loss:.4f}")
    print(f"Test Accuracy: {accuracy:.2f}%")

evaluate_model(model, test_dataloader, processor, device)
```
